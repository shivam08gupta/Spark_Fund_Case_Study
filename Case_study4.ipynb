```python
#Data Cleaning 1
#1. How many unique companies are present in rounds2?
rounds2['company_permalink']=rounds2['company_permalink'].str.upper()
rounds2['company_permalink']=rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
Total_rounds2=len(rounds2['company_permalink'].unique().tolist())
print(Total_rounds2)
```

```python
#3. In the companies data frame, which column can be used as the unique key for each company? Write the name of the column.
# ANSWER: permalink
if companies.permalink.nunique()==len(companies.permalink.index):
  print("permalink")
else:
  print("Duplicate")  
```

```python
# 4. Are there any companies in the rounds2 file which are not present in companies? Answer yes or no: Y/N
# change in upper case
rounds2['company_permalink']=rounds2['company_permalink'].str.upper()
companies['permalink']=companies['permalink'].str.upper()

rounds2.company_permalink.isin([x for x in np.unique(rounds2.company_permalink) if x not in companies.permalink])
a=rounds2.company_permalink.isin([x for x in np.unique(rounds2.company_permalink) if x not in companies.permalink])

# print(a)
if a.any():
 print("N")
else:
 print("Y")
```

```python
# 5. Merge the two data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame. 
# Name the merged frame master_frame. How many observations are present in master_frame?

master_frame = pd.merge(rounds2,companies,left_on='company_permalink',right_on='permalink', how='left')
# To check how many null value
nan_rows = master_frame[master_frame['raised_amount_usd'].isnull()]
print("Total Null Value" , len(nan_rows))

# master_frame = master_frame.dropna(axis=0, subset=['raised_amount_usd'])
# master_frame
master_frame['raised_amount_usd'] = master_frame['raised_amount_usd'].replace(np.nan, 0)
master_frame.shape
```

```python
# Sector Analysis 1
# code for importing csv files
mapping = pd.read_csv("C:/Users/shivamguptayo/Desktop/Data Science/Case Study/mapping.csv", sep=",", encoding="ISO-8859-1")
mapping
```

```python
# When we say sector analysis, we refer to one of the eight main sectors (named main_sector) listed in the mapping file 
# (note that ‘Other’ is one of the eight main sectors). This is to simplify the analysis by grouping the numerous 
# category lists (named ‘category_list’) in the mapping file. For example, in the mapping file, category_lists 
# such as ‘3D’, ‘3D Printing’, ‘3D Technology’, etc. are mapped to the main sector ‘Manufacturing’.

#  Name is id_vars and Course is value_vars 

# Checkpoint 4: Sector Analysis 1
# 2. Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors 
# (Note that ‘Others’ is also considered one of the main sectors)

df=pd.DataFrame()
df=pd.melt(mapping, id_vars =['category_list'], var_name ='main_sector')
main_sector_mapping=df.loc[df['value']==(1)]
main_sector_mapping
```

```python
# Checkpoint 4: Sector Analysis 1
# 1. Extract the primary sector of each category list from the category_list column

# the category list is a list of multiple sub-sectors separated by a pipe (vertical bar |).
# For example, one of the companies’ category_list is Application Platforms|Real Time|Social Network Media.


# master_frame=pd.DataFrame(master_frame)
new = master_frame["category_list"].str.split("|", n = 1, expand = True) 
master_frame["Primary_sector"]= new[0]
master_frame.head()
# master_frame.shape
```

```python
# Checkpoint 4: Sector Analysis 1
# 2. Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors
# (Note that ‘Others’ is also considered one of the main sectors)

mapping_sector = pd.merge(master_frame,main_sector_mapping,left_on='Primary_sector',right_on='category_list', how='left')
mapping_sector
```
